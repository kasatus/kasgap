```python
'''
returns tuple of 
(1) renewal ratio for given major line
(2) total number of elements
(3) number of not null elements in 'renewed yorn'
'''
def renewal_ratio(arr): 
    c = Counter(arr) 
    ratio = float(c['Y'])/(c['Y']+c['N'])
    total = len(arr)
    not_null = c['Y']+c['N']
    return (ratio, total, not_null)
```
